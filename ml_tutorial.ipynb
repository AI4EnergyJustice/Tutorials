## 2. Scikit-Learn Library <a name="scikit"></a>

[Scikit-learn](https://scikit-learn.org/stable/) is a powerful and versatile Python library for machine learning and data analysis. It provides a wide range of algorithms for supervised and unsupervised learning, as well as tools for model fitting, data preprocessing, model selection and evaluation, and many other utilities. Built on NumPy, SciPy, and matplotlib, scikit-learn's simple and consistent API makes it accessible to beginners, while its efficient implementations of a large number of algorithms and rich functionality make it valuable for experts. The library's comprehensive [documentation](https://scikit-learn.org/stable/user_guide.html) and active community make it an excellent resource for anyone looking to do machine learning with Python.

In this demo, we'll use the famous Iris dataset, which is included in Scikit-Learn's datasets module. We'll use a simple classifier, the K-Nearest Neighbors (KNN), to predict the species of Iris.

Here's the outline of the steps we'll take:

1. **Load the Dataset:** We'll start by loading the Iris dataset from Scikit-Learn's datasets module.
2. **Exploratory Data Analysis (EDA):** We'll take a look at the dataset to understand what it contains.
3. **Data Preprocessing:** We'll prepare the data for the machine learning model.
4. **Train the Model:** We'll use the K-Nearest Neighbors (KNN) algorithm to train our model.
5. **Evaluate the Model:** Finally, we'll evaluate the performance of our model using accuracy score and a confusion matrix.


```python

# Import necessary libraries
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Load Iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

```

The dataset consists of 150 samples from each of three species of Iris flowers (Iris setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the lengths and the widths of the sepals and petals.


```python
from IPython.display import Image, display

# Load images
img1 = Image(filename='images/640px-Irissetosa1.jpg', width=50)
img2 = Image(filename='images/480px-Blue_Flag,_Ottawa.jpg', width=50)
img3 = Image(filename='images/480px-Iris_virginica_2.jpg', width=50)

# Display images with names
# TODO: Use widgets or html to display images side by side
print("Iris Setosa")
display(img1)

print("Iris Versicolor")
display(img2)

print("Iris Virginica")
display(img3)
```


The columns in this DataFrame are:

* sepal length (cm): the length of the sepal
* sepal width (cm): the width of the sepal
* petal length (cm): the length of the petal
* petal width (cm): the width of the petal
* target: the species of the iris flower. This is what we'll predict. It's a number where 0 represents Iris-Setosa, 1 represents Iris-Versicolor, and 2 represents Iris-Virginica.

```python
# Display the first few rows of the DataFrame
print(iris_df.head())
```

```python
# Summary statistics
print("\nSummary statistics:")
print(iris_df.describe())
# The summary statistics give us a sense of the distribution of values in each feature.
```

```python

# Pairplot to visualize the distribution of each feature and the relationships between different features
sns.pairplot(iris_df, hue='target')
plt.show()
```

The pairplot visualizes the relationships between different features and how
they are distributed. Each point represents a flower, colored by its species.
You can see that some species are clearly separable based on certain features,
which indicates that our machine learning model should be able to learn to
distinguish them.

Next, we'll split our dataset into a training set and a test set. This allows us
to evaluate how our model performs on unseen data. We'll use 70% of the data for
training and 30% for testing. Then, we'll train our K-Nearest Neighbors (KNN)
model on the training data.



```python

# Split the data into training set and test set
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X_train, y_train)
```

The model has been trained using the K-Nearest Neighbors algorithm.

Let's now use our trained model to make predictions on the test set. We'll then evaluate the model's performance by calculating the accuracy and generating a confusion matrix. The [accuracy score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html) gives us a simple measure of how often the model makes correct predictions, and the [confusion matrix](https://www.ml-science.com/confusion-matrix) provides a breakdown of where the model's predictions go wrong.

```python

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display the results
print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Confusion Matrix:\n{cm}")
```

The model achieved an accuracy of 100% on the test set, which means it correctly predicted the species of every iris in the test set. This is a great result!

The confusion matrix further confirms this, as there are no off-diagonal elements, which represent incorrect predictions.

Please note that while 100% accuracy is great, in the real world, perfect accuracy is rare and can sometimes even be a sign of overfitting. Overfitting is when a model learns the training data too well, to the point where it performs poorly on unseen data. However, in this case, since we're achieving 100% accuracy on the test set, not the training set, overfitting is unlikely to be an issue.
